```
import numpy as np
import matplotlib.pyplot as plt
```

```
def reLu(x):
  return np.maximum(0,x)
```

```
x = np.arange(-2,2,0.2)
x
```

```
plt.axhline(0,color='grey')
plt.axvline(0,color='grey') 
plt.plot(x,reLu(x))
```

# Approximating x^3 + x^2 -x -1

```
x = np.arange(-2,2,0.01)
x
```

```
f=x**3+x**2-x-1
```

```
plt.plot(x,f)
```

```
f1=reLu(-5*x-7.7)
plt.plot(x,f)
plt.plot(x,f1)
```

```
f2=reLu(-1.2*x-1.3)
plt.plot(x,f)
plt.plot(x,f2)
```

```
f3=reLu(1.2*x+1)
plt.plot(x,f)
plt.plot(x,f3)
```

```
f4=reLu(1.2*x-2)
plt.plot(x,f)
plt.plot(x,f4)
```

```
f5=reLu(2*x-1.1)
plt.plot(x,f)
plt.plot(x,f5)
```

```
f6=reLu(5*x-5)
plt.plot(x,f)
plt.plot(x,f6)
```

```
plt.axhline(0,color='grey')
plt.axvline(0,color='grey') 
plt.plot(x,f)
plt.plot(x,-f1-f2-f3+f4+f5+f6)
```
