```
import cv2 
```

```
img = cv2.imread('krishna.jpg')
```

```
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```

```
img.shape
```

```
width,height = 300,300
imgresize = cv2.resize(img,(width,height),interpolation=cv2.INTER_CUBIC)
```

```
cv2_imshow(imgresize)
```

```
imgresize1 = cv2.resize(img,(width,height),interpolation=cv2.INTER_NEAREST)
cv2_imshow(imgresize1)
```

```
imgresize2 = cv2.resize(img,(width,height),interpolation=cv2.INTER_LINEAR)
cv2_imshow(imgresize2)
```

```
imgresize3 = cv2.resize(img,(width,height),interpolation=cv2.INTER_AREA)
cv2_imshow(imgresize3)
```

```
imgresize4 = cv2.resize(img,(width,height),interpolation=cv2.INTER_LINEAR_EXACT)
cv2_imshow(imgresize4)
```

```
width,height = 800,600
imgresize5 = cv2.resize(img,(width,height),interpolation=cv2.INTER_CUBIC)
cv2_imshow(imgresize5)
```

```
imgresize6 = cv2.resize(img,(width,height),interpolation=cv2.INTER_LINEAR_EXACT)
cv2_imshow(imgresize6)
```

```
imgresize7 = cv2.resize(img,(width,height),interpolation=cv2.INTER_LANCZOS4)
cv2_imshow(imgresize7)
```
