```
import numpy as np
import matplotlib.pyplot as plt
```

```
from matplotlib import cycler
colors = cycler('color',
                ['#EE6666', '#3388BB', '#9988DD',
                 '#EECC55', '#88BB44', '#FFBBBB'])
plt.rc('axes', facecolor='#E6E6E6', edgecolor='none',
       axisbelow=True, grid=True, prop_cycle=colors)
plt.rc('grid', color='w', linestyle='solid')
plt.rc('xtick', direction='out', color='gray')
plt.rc('ytick', direction='out', color='gray')
plt.rc('patch', edgecolor='#E6E6E6')
plt.rc('lines', linewidth=2)
```

```
def linear(x):
  return x 
```

```
x = np.arange(-2,2,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, linear(x))
```

```
def eLU(x, alpha=1):
  return np.array([xi if xi>0 else alpha*(np.exp(xi)-1) for xi in x ])
```

```
x=np.arange(-10,10,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, eLU(x))
```

```
# With a different alpha
x=np.arange(-2,2,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, eLU(x,5))
```

```
def logsigmoid(x, lam=1):
  return np.log(1/(1+np.exp(-lam*x)))
```

```
x=np.arange(-10,10,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, logsigmoid(x))
```

```
# With a different lambda
x=np.arange(-10,10,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, logsigmoid(x,5))
```

**Tanh**

```
def tanh(x, lam=1):
  return ((np.exp(lam*x)-np.exp(-lam*x))/(np.exp(lam*x)+np.exp(-lam*x)))
```

```
# With a different lambda
x=np.arange(-2,2,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, tanh(x))
```
